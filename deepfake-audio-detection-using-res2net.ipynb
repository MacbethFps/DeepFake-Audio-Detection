```python
!pip install resampy
import numpy as np
import pandas as pd
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
from tqdm import tqdm
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, BatchNormalization, Activation, 
    GlobalAveragePooling2D, Dense, Concatenate
)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler

# Enable mixed precision for lower memory usage
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy("mixed_float16")

# ------------------ Res2Net Block ------------------ #
def res2net_block(x, filters, scales=4, kernel_size=3, stride=1):
    split_channels = filters // scales
    y = []
    
    for i in range(scales):
        if i == 0:
            y.append(Conv2D(split_channels, (1, 1), strides=(1, 1), padding="same", activation="relu")(x))
        else:
            y.append(Conv2D(split_channels, (kernel_size, kernel_size), strides=(stride, stride), padding="same", activation="relu")(y[-1]))
    
    out = Concatenate(axis=-1)(y)
    out = Conv2D(filters, (1, 1), padding="same")(out)
    out = BatchNormalization()(out)
    
    return Activation("relu")(out)

# ------------------ Res2Net Model ------------------ #
def create_res2net(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    x = Conv2D(32, (3, 3), padding="same", activation="relu")(inputs)
    x = res2net_block(x, 64)
    x = res2net_block(x, 128)
    x = res2net_block(x, 256)
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation="relu")(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    model = Model(inputs, outputs)
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

# ------------------ Load & Preprocess Data ------------------ #
audio_files_path = "/kaggle/input/deep-voice-deepfake-voice-recognition/KAGGLE/AUDIO/"
folders = os.listdir(audio_files_path)

data, labels = [], []
max_duration = 5.0  # 🔹 Load only 5 seconds of audio

for folder in folders:
    files = os.listdir(os.path.join(audio_files_path, folder))[:100]  # 🔹 Limit to 100 files per class
    for file in tqdm(files):
        file_path = os.path.join(audio_files_path, folder, file)
        audio, sample_rate = librosa.load(file_path, res_type="kaiser_fast", duration=max_duration)
        
        # Convert audio to MFCC spectrogram
        mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        data.append(mfccs_features)
        labels.append(folder)

# Convert to DataFrame
feature_df = pd.DataFrame({"features": data, "class": labels})

# Encode labels
le = LabelEncoder()
feature_df["class"] = le.fit_transform(feature_df["class"])
num_classes = len(le.classes_)

# ------------------ Fix Shape Issue (Padding MFCCs) ------------------ #
max_length = max([mfcc.shape[1] for mfcc in feature_df["features"]])

def pad_mfcc(mfcc, max_len):
    if mfcc.shape[1] < max_len:
        pad_width = max_len - mfcc.shape[1]
        return np.pad(mfcc, ((0, 0), (0, pad_width)), mode="constant")
    else:
        return mfcc[:, :max_len]  # Truncate if too long

X = np.array([pad_mfcc(mfcc, max_length) for mfcc in feature_df["features"]])
y = np.array(feature_df["class"].tolist())

# Reshape for CNN input (add channel dimension)
X = np.expand_dims(X, axis=-1)

# ------------------ Data Balancing & Splitting ------------------ #
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X.reshape(len(X), -1), y)
X_resampled = X_resampled.reshape(len(X_resampled), 40, max_length, 1)  # Reshape back to image-like structure
y_resampled = to_categorical(y_resampled)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# ------------------ Train the Res2Net Model ------------------ #
input_shape = X_train.shape[1:]
model = create_res2net(input_shape, num_classes)
model.summary()

# Save model checkpoints to prevent losing progress
# Save model checkpoints to prevent losing progress
checkpoint = ModelCheckpoint("res2net_model.keras", save_best_only=True, monitor="val_loss", mode="min")


early_stop = EarlyStopping(monitor="val_loss", patience=5)
history = model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test), 
    batch_size=4,  # 🔹 Reduce batch size to avoid crashes
    epochs=2,      # 🔹 Reduce epochs
    callbacks=[early_stop, checkpoint]
)

# ------------------ Evaluate Model ------------------ #
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc)

# ------------------ Plot Training History ------------------ #
plt.figure()
plt.title("Model Accuracy")
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="validation")
plt.legend()
plt.ylim([0, 1])
plt.show()

plt.figure()
plt.title("Model Loss")
plt.plot(history.history["loss"], label="train")
plt.plot(history.history["val_loss"], label="validation")
plt.legend()
plt.ylim([0, 1])
plt.show()

# ------------------ Fake Audio Detection Function ------------------ #
def detect_fake(filename):
    sound_signal, sample_rate = librosa.load(filename, res_type="kaiser_fast", duration=max_duration)
    mfcc_features = librosa.feature.mfcc(y=sound_signal, sr=sample_rate, n_mfcc=40)
    mfcc_features = pad_mfcc(mfcc_features, max_length)  # Ensure consistent shape
    mfcc_features = np.expand_dims(mfcc_features, axis=-1)
    mfcc_features = np.expand_dims(mfcc_features, axis=0)  # Reshape for model
    
    result_array = model.predict(mfcc_features)
    result_classes = ["FAKE", "REAL"]
    result = np.argmax(result_array[0])
    print("Result:", result_classes[result])

# ------------------ Test Fake Audio Detection ------------------ #
test_real = "/kaggle/input/deep-voice-deepfake-voice-recognition/DEMONSTRATION/DEMONSTRATION/linus-original-DEMO.mp3"
test_fake = "/kaggle/input/deep-voice-deepfake-voice-recognition/DEMONSTRATION/DEMONSTRATION/linus-to-musk-DEMO.mp3"

detect_fake(test_real)
detect_fake(test_fake)
```
